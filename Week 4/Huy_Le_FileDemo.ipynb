## Lambda Functions

Lambda functions are **small, anonymous functions** written in a single line.

**Syntax:** `lambda arguments: expression`

- Used for **simple operations**
- Always return a value
- No function name is required

```python
# Example 1: addition
add = lambda a, b: a + b
add(3, 7)
```

```python
# Example 2: check if a number is even
is_even = lambda x: x % 2 == 0
is_even(10)
```

## Lists

Lists are **ordered, mutable collections**.

They allow:
- Duplicate values
- Changing values after creation

A list is used to store multiple values in one place when the data can change and order matters.

```python
# Creating a list
numbers = [1, 2, 3, 4]
numbers
```

```python
# Accessing elements
numbers[0]
```

```python
# Modifying a list
numbers[2] = 10
numbers
```

```python
# Print the list
for number in numbers:
    print(number)
```

## Tuples

Tuples are **ordered but immutable**.

Once created, their values **cannot be changed**.

Tuples are used to store multiple values that should not be changed, ensuring data stays safe and consistent.

```python
# Creating a tuple
days = ('Mon', 'Tue', 'Wed')
days
```

```python
# Accessing tuple values
days[1]
```

## Sets

Sets store **unique values only** and have **no fixed order**.

Sets are used to store multiple values when you want only unique items and the order does not matter.

```python
# Creating a set
unique_numbers = {1, 2, 2, 3, 4}
unique_numbers
```

```python
# Adding elements to a set
unique_numbers.add(5)
unique_numbers
```

```python
first_element = next(iter(unique_numbers))
print(first_element)
```

```python
# Covert to List
unique_numbers_list = list(unique_numbers)
print(unique_numbers_list[0])
```

```python
# Remove form set
unique_numbers_pop = unique_numbers.pop()
print(unique_numbers_pop)
print(type(unique_numbers_pop))
```

## Dictionaries

Dictionaries store data as **key–value pairs**.

Each key must be unique.

Dictionaries are used to store data as key–value pairs so values can be quickly accessed using meaningful keys.

```python
# Creating a dictionary
student = {
    "name": "Sara",
    "age": 19,
    "major": "Computer Science"
}
student
```

```python
# Accessing dictionary values
student['major']
```

## Local and Global Scope

- **Global variables** are defined outside functions
- **Local variables** exist only inside a function

```python
# Global variable
x = 50

def show_x():
    print('Inside function:', x)

show_x()
print('Outside function:', x)
```

```python
# Local variable example
def local_scope():
    y = 20
    return y

local_scope()
```
